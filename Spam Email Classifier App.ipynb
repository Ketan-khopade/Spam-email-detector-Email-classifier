```python
from tkinter import Tk,Label,messagebox
from tkinter.ttk import Button,Entry,Style
import pickle
import numpy as np
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
import string
ps = PorterStemmer()

def remove_special_characters(x):
    text = ''
    for i in x:
        if i in string.punctuation:
            continue
        else:
            text += i
    return text.lower()
def stopword(x):
    y = []
    for j in x.split():
        if j in stopwords.words('english'):
            continue
        else:
            y.append(ps.stem(j))
    return ' '.join(y)

def output(text):
    tfidf = pickle.load(open('vector.pkl','rb'))
    model = pickle.load(open('model.pkl','rb'))
    text1 = remove_special_characters(text)
    text2 = stopword(text1)
    data = tfidf.transform([text2]).toarray()
    pred = model.predict(data)
    root = Tk()
    root.geometry('400x275')
    root.maxsize(width=300, height=200)
    root.minsize(width=300, height=200)
    root.title('spam email predictor')
    root.config(bg='LightSteelBlue1')
    style2 = Style()
    style2.configure('TButton', bg='LightSkyBlue3', font=('Franklin Gothic Demi', 10, 'bold'))
    label = Label(text='spam' if pred ==1 else 'not spam')
    label.place(relx=0.35,rely=0.4)
    button = Button(text='EXIT',command=root.destroy)
    button.place(relx=0.35,rely=0.6)
    root.mainloop()
def predict():
    root = Tk()
    root.geometry('400x275')
    root.maxsize(width=300, height=200)
    root.minsize(width=300, height=200)
    root.title('spam email predictor')
    root.config(bg='LightSteelBlue1')
    style2 = Style()
    style2.configure('TButton', bg='LightSkyBlue3', font=('Franklin Gothic Demi', 10, 'bold'))
    def clear(event):
        entry.delete(0,20)
        
    entry = Entry(width=40)
    entry.insert(0,'Enter a Text')
    entry.bind('<Button-1>',clear)
    
    def predict1():
        if entry.get() == 'Enter a Text':
            messagebox.showinfo('enter please')
        else:
            data = entry.get().strip()
            root.destroy()
            output(data)
    button = Button(text='EXIT',command=root.destroy).pack(anchor='ne',side='top')
    entry.place(relx=0.1,rely=0.4)
    button1 = Button(text = 'PREDICT',command=predict1)
    button1.place(relx=0.35,rely=0.6)
    root.mainloop()
    
predict()
```
